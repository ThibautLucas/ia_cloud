```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
rand=[]
nbr_random = 100000
nbr_array = 1000
for x in range(nbr_array):
   rand.append([])
   for i in range(nbr_random):
        rand[x].append(np.random.randn())
print(len(rand))
    
```

```python
for row in rand:
    counts, bins = np.histogram(row)
    plt.hist(bins[:-1], weights=counts, bins=bins)
```

```python
mean_array=[]
for x in rand:
   mean_array.append(np.histogram(row)[0])
means = np.mean(mean_array, axis=0)
plt.plot(means)
```
