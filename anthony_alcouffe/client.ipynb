```python
import urllib.request
from matplotlib import pyplot as plt
import json
import cv2
```

```python
image_path = "images/face-2.jpg"
contents = urllib.request.urlopen("http://127.0.0.1:5000/predict?name=" + image_path).read().decode("utf-8")
```

```python
json_data = json.loads(contents)
image = cv2.cvtColor(cv2.imread(image_path), cv2.COLOR_BGR2RGB)
```

```python
color = (255, 0, 0) 
thickness = 2
for row in json_data:
    cv2.rectangle(image, (row["x"], row["y"]),
      (row["x"] + row["h"], row["y"] + row["w"]),
      (255,0,0), thickness)
plt.imshow(image)
plt.show()

```
